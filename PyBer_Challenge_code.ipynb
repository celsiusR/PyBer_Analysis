### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
#The total number of rides for each city type is retrieved
total_rides_byCityType = pyber_data_df.groupby(["type"]).count()["ride_id"]
#total_rides_byCityType

total_rides_byCityType.groupby(['type']).sum().plot(kind='pie', autopct="%1.1f%%", ylabel="RIDES")
```

```python
# 2. Get the total drivers for each city type
#The total number of drivers for each city type is retrieved
total_drivers_byCityType = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_byCityType
total_drivers_byCityType.groupby(['type']).sum().plot(kind='pie', autopct="%1.1f%%", ylabel="DRIVERS")
```

```python
#  3. Get the total amount of fares for each city type
#The sum of the fares for each city type is retrieved
total_fares_byCityType = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_byCityType
total_fares_byCityType.groupby(['type']).sum().plot(kind='pie', autopct="%1.1f%%", ylabel="TOTAL FARES")
```

```python
#  4. Get the average fare per ride for each city type. 
#The average fare per ride for each city type is calculated
average_fares_per_ride = total_fares_byCityType/total_rides_byCityType
average_fares_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
#The average fare per driver for each city type is calculated
average_fare_per_driver= total_fares_byCityType/total_drivers_byCityType
```

```python
#  6. Create a PyBer summary DataFrame. 
#A PyBer summary DataFrame is created
PyBerSummary_df=pd.DataFrame()
PyBerSummary_df ["Total Rides"] = total_rides_byCityType.map("{:,}".format)
PyBerSummary_df ["Total Drivers"] = total_drivers_byCityType.map("{:,}".format)
PyBerSummary_df ["Total Fares"] = total_fares_byCityType.map("${:,.2f}".format)
PyBerSummary_df ["Average Fare per Ride"] = average_fares_per_ride.map("${:.2f}".format)
PyBerSummary_df ["Average Fare per Driver"] = average_fare_per_driver.map("${:.2f}".format)
PyBerSummary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
#The PyBer summary DataFrame is formatted as shown in the example
PyBerSummary_df.index.name = None
PyBerSummary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

new_sum_of_fares = pyber_data_df.groupby(["type", "date"]).sum()[["fare"]]
new_sum_of_fares
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
new_sum_of_fares = new_sum_of_fares.reset_index()

```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
new_sum_fares_pivot = new_sum_of_fares.pivot(index="date", columns="type")["fare"]
new_sum_fares_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
new_pivot_df = new_sum_fares_pivot.loc["2019-01-01":"2019-04-28"]
new_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
new_pivot_df.index = pd.to_datetime(new_pivot_df.index)

new_pivot_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
new_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
final_by_week_df = new_pivot_df.resample("W").sum()
final_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
final_by_week_df.plot(figsize=(20,8))

plt.gcf().subplots_adjust(bottom=0.15)

# Add graph properties
plt.title("Total Fare by City Type", fontsize=16)
plt.ylabel("Fare ($USD)", fontsize=14)
plt.xlabel("Month", fontsize=14)

# Create a legend
# from an aesthetic perspective I would have preferred to use 
# lgnd = plt.legend(fontsize="10", loc="best", title="City Type")
# But adhering to the Requirements worked it out as loc="center"
lgnd = plt.legend(fontsize="10", loc="center", title="City Type")
lgnd.get_title().set_fontsize(12)

# Save figure
plt.savefig("Analysis/total_fares_by_city_type.png")

# Show figure
plt.show()

```
